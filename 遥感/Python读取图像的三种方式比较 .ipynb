## 实际操作

### Pillow

```python
from PIL import Image
image_pil=Image.open("../Assets/kid.tif")
r,g,b=image_pil.split()
r.show()
```

### matplotlib

> Matplotlib relies on the Pillow library to load image data.
>
> matplotlib 依赖Pillow库以加载图像数据
>
> [Matplotlib Image tutorial](https://matplotlib.org/stable/tutorials/introductory/images.html)

matplotlib 读进来的图像是RGB顺序，

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
image=mpimg.imread("../Assets/kid.tif")
print(f'变量类型：{type(image)}\n图像形状：{image.shape}\n图像数据类型：{image.dtype}')
fig, (ax0, ax1, ax2,ax3) = plt.subplots(1,4,figsize=(15, 6))

ax0.set_title("origin")
ax0.imshow(image)

ax1.set_title("R Band")
pos1=ax1.imshow(image[:,:,0],cmap="gray")

ax2.set_title("G Band")
pos2=ax2.imshow(image[:,:,1],cmap="gray")

ax3.set_title("B Band")
pos3=ax3.imshow(image[:,:,2],cmap="gray")
fig.colorbar(pos3,ax=[ax1,ax2,ax3],orientation='horizontal')
```

### OpenCV

```python
import cv2 as cv
image_cv=cv.imread("../Assets/kid.tif")
image_cv_to_rgb=cv.cvtColor(image_cv,cv.COLOR_BGR2RGB)
print(f'变量类型：{type(image_cv)}\n图像形状：{image_cv.shape}\n图像数据类型：{image_cv.dtype}')
plt.imshow(image_cv[:,:,0])
```
